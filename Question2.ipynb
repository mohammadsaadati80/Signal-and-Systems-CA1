```python
import numpy
from matplotlib import pyplot
from scipy import signal
from scipy import ndimage
impulse = signal.unit_impulse(100, [0])
n = np.arange(0,100,1)
```

#### Impulse response

#### A.

```python
pyplot.stem(n , (0.5) * (ndimage.shift(impulse, 0, cval=0)[n] + (ndimage.shift(impulse, 1, cval=0)[n])))
pyplot.ylabel('y[n]')
pyplot.xlabel('n')
pyplot.title('Y1[n]')
pyplot.grid()
pyplot.show()
```

#### B.

```python
pyplot.stem(n , (ndimage.shift(impulse, -1, cval=0)[n] - ndimage.shift(impulse, 0, cval=0)[n]))
pyplot.ylabel('y[n]')
pyplot.xlabel('n')
pyplot.title('Y2[n]')
pyplot.grid()
pyplot.show()
```

#### C.

```python
pyplot.stem(n , (ndimage.shift(impulse, -1, cval=0)[n] - 2 * ndimage.shift(impulse, 0, cval=0)[n] + ndimage.shift(impulse, 1, cval=0)[n]))
pyplot.ylabel('y[n]')
pyplot.xlabel('n')
pyplot.title('Y3[n]')
pyplot.grid()
pyplot.show()
```

### Part 2

```python
def my_sum(inp, bound):
    _sum = 0
    result = []
    for i in range(len(inp)):
        if i >= bound:
            _sum = 0
        else:
            _sum += inp[i]
        result.append(_sum)
    return result
```

#### A.

```python
pyplot.stem(n , my_sum((0.5) * (ndimage.shift(impulse, 0, cval=0)[n] + (ndimage.shift(impulse, 1, cval=0)[n])), 5))
pyplot.ylabel('y[n]')
pyplot.xlabel('n')
pyplot.title('Step Response Y1[n] m_u = 5')
pyplot.grid()
pyplot.show()
```
