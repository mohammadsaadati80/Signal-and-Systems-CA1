### Part A

```python
import numpy as np
from matplotlib import pyplot as plt

def x(A, n, k):
    x = A*(np.cos((2*np.pi)* k * n))
    return x

def X(n):
    y = x(A, n, k)
    plt.stem(n, y)
    plt.title('X[n]')
    plt.xlabel('n')
    plt.ylabel('X[n]')
    plt.show()
    
def Y1(n):
    y = (1/T)*(x(A, n+1, k) - x(A, n, k))
    plt.stem(n, y)
    plt.title('Y1[n]')
    plt.xlabel('n')
    plt.ylabel('Y[n]')
    plt.show()
    
def Y2(n):
    y = (1/(2*T))*(x(A, n+1, k) - x(A, n-1, k))
    plt.stem(n, y)
    plt.title('Y2[n]')
    plt.xlabel('n')
    plt.ylabel('Y[n]')
    plt.show()

def Y3(n):
    y = (1/(T**2))*(x(A, n+1, k) - (2)*x(A, n, k) + x(A, n-1, k))
    plt.stem(n, y)
    plt.title('Y3[n]')
    plt.xlabel('n')
    plt.ylabel('Y[n]')
    plt.show()
    
def Y4(n):
    y = (0.1)*x(A, n-2, k) + (0.2)*x(A, n-1, k) + (0.4)*x(A, n, k) + (0.2)*x(A, n+1, k) + (0.1)*x(A, n+2, k) 
    plt.stem(n, y)
    plt.title('Y4[n]')
    plt.xlabel('n')
    plt.ylabel('Y[n]')
    plt.show()
    
A = 2
n = np.array([n for n in range(0, 16)])
k = 0.8
T = 1 / k

X(n)
```
